```python
# Import Library
from PIL import Image, ImageDraw, ImageFont
import matplotlib.pyplot as plt
```

```python
# Membuka citra
image = Image.open("/content/catterpillar (8).jpg")

# Menampilkan Citra
plt.subplot(1,2,1)
plt.title("Original Image")
plt.imshow(image)
```

```python
# Proses Watermarking Text Citra
watermark_image = image.copy()

draw = ImageDraw.Draw(watermark_image)
w,h = image.size
x,y = int(w/2), int(h/2)
if x > y:
    font_size = y
elif y > x:
    font_size = x
else:
    font_size = x

font = ImageFont.truetype("/content/Arial.ttf", int(font_size/3))

draw.text((x,y), "Yuhu", font=font, anchor='ms')
```

```python
# Menampilkan Hasil Watermark Text
plt.subplot(1,2,2)
plt.title("Image with Watermark")
plt.imshow(watermark_image)
plt.show()
```

```python
# Penerapan Watermarking Image
size = (200, 50)
watermark_image = Image.open("/content/catterpillar (30).jpg")

# Menentukan ukuran Watermark Citra
watermark_image.thumbnail(size)

copied_image = image.copy()
Image.Image.paste(copied_image,  watermark_image, (130,80))

plt.title("Image with Watermark")
plt.imshow(copied_image)
plt.show()
```
